```python
import librosa

import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

## Important Commands

np.linspace(start, end, n_samples) -> evenly spacing in {start, end}<br>
np.arange(N) -> makes an array with the elements [0,1,2, .... N-1]

```python
sr = 44100 # sample rate

#T = 440    # Period

f = 440   # 440 cycles per second
time = 2  # seconds of auidio
phi = 1   # Phase



#t = np.linspace(0,100, 1000) # space 1000 samples evenly in [0,100]
t = np.arange(sr*time) / sr # in units of seconds
y = np.cos(2*np.pi*f*t + phi)
plt.plot(t,y)

ipd.Audio(y, rate=sr)
```
