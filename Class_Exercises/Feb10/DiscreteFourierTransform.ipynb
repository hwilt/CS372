```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.signal import fftconvolve
import librosa
```

```python
def get_dft(x):
    N = len(x)
    n_freqs = int(np.ceil(N/2))
    cos_sums = np.zeros(n_freqs)
    sin_sums = np.zeros(n_freqs)
    
    ## TODO: Fill this in
    ts = np.arange(N)/N
    for f in range(n_freqs):
        c = np.cos(2*np.pi*f*ts)
        s = np.sin(2*np.pi*f*ts)
        cos_sums[f] = np.dot(c,x)
        sin_sums[f] = np.dot(s,x)
        #for n in range(N):
        #    cos_sums[k] += x[n]*np.cos(2*np.pi*k*n/N)
        #    sin_sums[k] += x[n]*np.sin(2*np.pi*k*n/N)

    return cos_sums, sin_sums
```

```python
N = 100
t = np.linspace(0, 1, N+1)[0:N] # Evenly spaced on [0, 1)
x = np.cos(2*np.pi*t)
c, s = get_dft(x)
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.stem(c)
plt.title("Cosine Product Sums") 
plt.subplot(122)
plt.stem(s)
plt.title("Sine Product Sums")
```
