```python
x = np.linspace(0, 1, 200)

A = np.zeros((100, 200)) # 100 rows, 200 columns
A[40, :] = x
# these are the same thing, [x][y] and [x, y]
A[50][50] = 1
A[20, 20] = 1

A[10:20, 50:75] = 2
A[80:90:2, 100:150:3] = 1

plt.imshow(A, interpolation='none') # interpolation='none' means no blurry images basically
plt.colorbar()
```

```python
N = 100
t = np.linspace(0, 2*np.pi, N)
f = 2
y = np.sign(np.cos(2*np.pi*f*t))

# TODO: Create 2D NxN Arrays C and S
C = np.zeros((N, N))
S = np.zeros((N, N))

n = np.arange(N)

for k in range(N):
    C[k, :] = np.cos(2*np.pi*k*n/N)
    S[k, :] = np.sin(2*np.pi*k*n/N)

c = C.dot(y) # Do the dot product of every row of C with y
s = S.dot(y) # Do the dot product of every row of S with y
 
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.stem(c)
plt.title("Cosine Product Sums") 
plt.subplot(122)
plt.stem(s)
plt.title("Sine Product Sums")
```

```python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.imshow(C, interpolation='none') # interpolation='none' means no blurry images basically
plt.colorbar()
plt.subplot(132)
plt.imshow(S, interpolation='none') # interpolation='none' means no blurry images basically
plt.colorbar()
```
