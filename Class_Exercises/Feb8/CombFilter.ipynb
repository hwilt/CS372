```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.signal import fftconvolve
import librosa
```

# Exercise 1: Shaping Noise

```python
sr = 44100
#ts = np.arange(sr*2)/sr

n = np.random.randn(sr*2)
h = np.zeros(1000)

T = 100
h[::T] = 1
plt.plot(h)
```

```python
y = fftconvolve(n, h)
plt.plot(y)
ipd.Audio(y, rate=sr)
```
