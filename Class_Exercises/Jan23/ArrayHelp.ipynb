```python
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
def get_cosine(time, sr, f):
    t = np.arange(int(sr*time))/sr
    return np.cos(2*np.pi*f*t)

f1 = 440
diff = 100
f2 = f + diff

sr = 44100
y1 = get_cosine(1.5, sr, f1)
y2 = get_cosine(1.5, sr, f2)
y = np.zeros(len(y1)+len(y2)) # big enough to hold both
y[0:len(y1)] = y1
y[len(y2)::] = y2
ipd.Audio(y, rate=sr)
```
