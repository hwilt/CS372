```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
n = np.arange(N)
y = np.zeros(N)
y += np.cos(2*np.pi*3*n/N)
y += 0.5*np.cos(2*np.pi*4*n/N)
y += 2*np.cos(2*np.pi*2*n/N)

# y = np.arange(N)

plt.plot(y)
plt.legend(["cos"])
```

```python
F = np.fft.fft(y)
plt.stem(np.abs(F))
#plt.xlim(-1,10)
plt.title(r"Amplitude")
plt.figure(figsize=[10,5])
plt.subplot(121)
plt.stem(np.real(F))
plt.xlim(0,10)
plt.title(r"Real (Cosine)")
plt.subplot(122)
plt.stem(np.imag(F))
plt.xlim(0,10)
plt.title(r"Imaginary (Sine)")
```
